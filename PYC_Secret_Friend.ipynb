```python
import pandas as pd
```

```python
import random
```

```python
file_path = "Desktop/PYC Secret Friend.xlsx"
```

```python
df = pd.read_excel(file_path)
```

```python
df.head(59)
```

```python
if 'gender' not in df.columns or 'name' not in df.columns:
    raise ValueError("The file must contain 'Name' and 'Gender' columns.")
```

```python
male = df[df['gender'] == 'male']['name'].tolist()
```

```python
female = df[df['gender'] == 'female']['name'].tolist()
```

```python
random.shuffle(male)
```

```python
random.shuffle(female)
```

```python
pairs = list(zip(male, female))
```

```python
extra_male = male[len(female):] if len(male) > len(female) else []
```

```python
extra_female = female[len(male):] if len(female) > len(male) else []
```

```python
print("\nPaired Names:")
for male, female in pairs:
    print(f"{male} - {female}")
```

```python
if extra_male:
    print("\nUnpaired Male:", extra_male)
```

```python
if extra_female:
    print("\nUnpaired Female:", extra_female)
```
